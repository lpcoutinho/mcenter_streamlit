```python
# Converte as colunas para o formato de tempo
df['hora_ini'] = pd.to_datetime(df['hora_ini'], format='%H:%M:%S')
df['hora_fim'] = pd.to_datetime(df['hora_fim'], format='%H:%M:%S')

# Calcula diferença entre as colunas 'hora_ini' e 'hora_fim'
df['tempo_total'] = df['hora_fim'] - df['hora_ini']

# Função para formatar a saída
def formatar_tempo(total):
    horas, restante = divmod(total.total_seconds(), 3600)
    minutos, segundos = divmod(restante, 60)
    return f'{int(horas):02d}:{int(minutos):02d}:{int(segundos):02d}'

# Aplica função à coluna 'tempo_total' para formatar o tempo
df['tempo_total'] = df['tempo_total'].apply(formatar_tempo)

# Calcula o tempo total da coluna 'tempo_total'
tempo_total = df['tempo_total'].apply(lambda x: pd.to_timedelta(x)).sum()

# Cria função para formatar o tempo total
def formatar_tempo_total(total):
    horas, restante = divmod(total.total_seconds(), 3600)
    minutos, segundos = divmod(restante, 60)
    return f'Total de horas trabalhadas: {int(horas):02d}:{int(minutos):02d}:{int(segundos):02d}'

# Exiba o total de horas trabalhadas
print(formatar_tempo_total(tempo_total))

# Valor da hora
valor_hora = 30  # Em reais

# Converta o tempo total para horas (em float)
tempo_total_em_horas = tempo_total.total_seconds() / 3600

# Calcule o valor total
valor_total = valor_hora * tempo_total_em_horas

# Exiba o valor total
print(f'Valor total: R${valor_total:.2f}')

```
